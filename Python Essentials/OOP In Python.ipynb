## Task 1: Student Profile
- Create a `Student` class that stores a name, age, and list of grades.
- Write an `__init__` method to set these attributes and add a `describe()` method that prints the student's details in a friendly sentence.
- Decide how the class should behave if an empty list of grades is provided.

Example usage:
```python
student = Student("Maya", 19, [14, 16, 18])
student.describe()
```

## Task 2: Average Grade Helper
- Extend your `Student` class with a method that returns the average of the stored grades.
- Think about what the method should return when the grade list is empty (number, message, `None`, etc.).
- Make sure the method does not change the original grades.

Example interaction:
```python
student = Student("Adel", 21, [12, 18, 15])
print(student.average_grade())  # Expected to show a numeric average
```

## Task 3: Rectangle Toolkit
- Build a `Rectangle` class with `width` and `height` attributes.
- Add methods that compute the area and perimeter.
- Include a method (or special method) that compares two rectangles and returns which one has a larger area.

Example prompt:
```python
rect_a = Rectangle(5, 3)
rect_b = Rectangle(4, 4)
print(rect_a.area())
print(rect_a.compare_by_area(rect_b))  # Decide what the return value looks like
```

## Task 4: Simple Bank Account
- Implement a `BankAccount` class with an owner name and starting balance.
- Provide `deposit(amount)` and `withdraw(amount)` methods that update the balance responsibly.
- Prevent the balance from dropping below zero and communicate why a withdrawal fails.

Example run:
```python
account = BankAccount("Sara", 100)
account.deposit(40)
account.withdraw(200)  # Should not allow overdraft
print(account.balance)
```

## Task 5: Playlist Manager

## Task 7: Vehicles and Cars

Example exploration:

## Task 8: Points in 2D
- Create a `Point` class that stores `x` and `y` coordinates.
- Implement `__str__` (or `__repr__`) for a readable display like `Point(2, 5)`.
- Implement `__eq__` so two points can be compared for equality.

Example check:
```python
p1 = Point(2, 5)
p2 = Point(2, 5)
p3 = Point(-1, 0)
print(p1)          # Should use your custom __str__
print(p1 == p2)    # Expected True
print(p1 == p3)    # Expected False
```

### Task 1: Import Libraries
Import `BeautifulSoup` from the `bs4` library.


### Task 2: Load Data
Read the content of `fake_bookstore.html` using Python's built-in `open()` function and store it in a variable named `html_content`.

> **Tip:** Use the `with open('filename.html', 'r') as f:` pattern to ensure the file closes automatically.

### Task 3: Make Soup
Create a `BeautifulSoup` object by passing your `html_content` and the parser string `'html.parser'`.

> **Tip:** `soup = BeautifulSoup(content, 'html.parser')`

### Task 4: Find by ID
Use the `.find()` method to extract the text of the main site title. The title has the id `site-title`.

> **Tip:** You can pass `id="site-title"` to `.find()`. Remember to access `.text` to get the string inside the tag.

### Task 5: Find by Class
Use `.find_all()` to get a list of all book titles. They all have the class `book-title`. Loop through the list and print each title's text.

> **Tip:** `class` is a reserved keyword in Python, so BeautifulSoup uses `class_="value"`.

### Challenge: Extract Prices
Extract the prices (class `price`), print them, and try to calculate their sum.

> **Tip:** The prices contain a `$` symbol. You might need to use string slicing (e.g., `text[1:]`) or `.replace()` before converting to a float.

## Step 1: Setup and Imports
First, we need to import our tools.

```python
import requests
from bs4 import BeautifulSoup
```

## Step 2: Fetching the Webpage
We use `requests.get()` to fetch the page. It returns a "Response" object.
We should always check the `.status_code`. A code of `200` means success!

```python
url = "http://books.toscrape.com/"
response = requests.get(url)

print(f"Status Code: {response.status_code}")

# If successful, the HTML content is stored in response.content or response.text
```

## Step 3: Making the "Soup"
Now we pass the HTML content to BeautifulSoup. This creates a tree structure that we can easily search.

```python
soup = BeautifulSoup(response.content, "html.parser")

# Let's print the page title to verify we have the right page
page_title = soup.title.text.strip()
print(f"Page Title: {page_title}")
```

## Example 1: Finding Elements by Tag Name (Headers)
The simplest way to find an element is by its tag name (like `h1`, `p`, `div`).
Let's find the main header of the page.

```python
# .find() returns the FIRST matching element it finds
header = soup.find("div", class_="page-header") # Finding a div with a specific class
if header:
    # Inside that div, let's find the h1 tag
    h1_tag = header.find("h1")
    print(f"Main Header: {h1_tag.text}")
```

## Example 2: Finding Elements by Class (Book Titles)
HTML elements often have `class` attributes to style them. We can use these to be specific.
On this website, every book is inside an `<article>` tag with class `product_pod`.
Inside that, the title is inside an `<h3>` tag, and inside that is an `<a>` tag.

```python
# Let's find the FIRST book to practice
first_book = soup.find("article", class_="product_pod")

# Navigate down to the title: article -> h3 -> a
book_title = first_book.h3.a["title"] # We get the 'title' attribute because the text inside might be truncated
print(f"First Book Title: {book_title}")
```

## Example 3: Finding Elements by Class (Prices)
Prices are usually distinct. Inspecting the page shows they are in a `<p>` tag with class `price_color`.

```python
price_tag = first_book.find("p", class_="price_color")
price = price_tag.text
print(f"Price: {price}")
```

## Example 4: Extracting All Books (Looping)
Usually, we want ALL the data, not just one item. 
We use `.find_all()` which returns a list of elements. We can then loop through them.

```python
# Find ALL articles with class 'product_pod'
all_books = soup.find_all("article", class_="product_pod")

print(f"Found {len(all_books)} books on this page.\n")

for book in all_books:
    # Extract title
    title = book.h3.a["title"]
    
    # Extract price
    price = book.find("p", class_="price_color").text
    
    print(f"{title} - {price}")
```

## Example 5: Handling Attributes (Links)
Sometimes the data isn't in the text, but in an attribute (like `href` for links or `src` for images).
We access attributes like a dictionary: `element["attribute_name"]`.

```python
for book in all_books[:3]: # Just the first 3 for brevity
    link_tag = book.h3.a
    relative_link = link_tag["href"]
    full_link = "http://books.toscrape.com/" + relative_link
    print(f"Link: {full_link}")
```

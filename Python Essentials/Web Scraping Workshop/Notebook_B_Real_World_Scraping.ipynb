### Task 1: Import Libraries
Import both `requests` and `BeautifulSoup`.

### Task 2: Request the Page
Use `requests.get()` to fetch the URL. Store the result in a variable named `response`.

> **Tip:** `response = requests.get(url)`

### Task 3: Check Status
Print the `status_code` of the response to ensure the request was successful. We are looking for code `200`.

> **Tip:** Access `.status_code` on your response object.

### Task 4: Make Soup
Create a `BeautifulSoup` object. This time, pass `response.content` instead of a file object.

> **Tip:** `BeautifulSoup(response.content, 'html.parser')`

### Task 5: Get Title
Extract the main title of the article. On Wikipedia, this is usually an `h1` tag with the id `firstHeading`.

> **Tip:** Use `.find(id="...")`.

### Task 6: Get Context
Extract the text of the first paragraph of the article.

> **Tip:** The main content is often in a div with class `mw-parser-output`. You might need to find that div first, then find the first `p` tag inside it.

### Task 7: Get Links
Find all `a` tags (links) in the "Table of Contents" and print the text of the first 5 links.

> **Tip:** The Table of Contents usually has an id of `toc` or class `toc`. Find that element first, then find all `a` tags within it. Use list slicing `[:5]` to get only the first five.

```python
# Conditionals: if/elif/else chain
temperature = 28

if temperature > 30:
    print("It's a hot day. Stay hydrated!")
elif temperature >= 20:
    print("The weather is perfect for a walk.")
else:
    print("It's chilly. Grab a jacket!")
```

```python
# For loop: repeating an action for each item
countdown = 3
for i in range(countdown, 0, -1):
    print(f"Countdown: {i}")
```

```python
# While loop: repeat until a condition changes
countdown = 3
while countdown > 0:
    print(f"Starting in {countdown}...")
    countdown -= 1
print("Go!")
```

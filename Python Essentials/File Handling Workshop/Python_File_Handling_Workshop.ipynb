```python
# RUN THIS CELL FIRST to generate the necessary practice files.
import json
import csv

# 1. Create story.txt
with open('story.txt', 'w') as f:
    f.write("Once upon a time in a land of code.\n")
    f.write("There lived a Python script.\n")
    f.write("It loved to process data every day.\n")

# 2. Create scores.csv
with open('scores.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow([10, 20, 30])
    writer.writerow([15, 25, 35])
    writer.writerow([12, 22, 32])

# 3. Create user_data.json
data = {
    "name": "Alice",
    "role": "Developer",
    "id": 101
}
with open('user_data.json', 'w') as f:
    json.dump(data, f)

print("Setup complete! Practice files created: story.txt, scores.csv, user_data.json")
```

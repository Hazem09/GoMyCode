## Basic Functions
Functions let you bundle reusable logic with a clear name and parameters.
Use `def` followed by the function name and parentheses to define them.

```python
def greet():
    print("Hello there!")

greet()  # Hello there!
```

## Functions with Multiple Arguments
Functions can accept positional parameters separated by commas, letting you operate on several inputs at once.
Order matters when calling a function with positional arguments.

```python
def add_numbers(a, b, c):
    print(a + b + c)

add_numbers(2, 3, 5)  # 10
```

## Default Arguments
Provide fallback values for parameters by assigning defaults in the function definition.
Callers can override defaults by passing explicit values.

```python
def power(base, exponent=2):
    return base ** exponent

print(power(4))      # 16
print(power(4, 3))   # 64
```

## Return Statement
Use `return` to send a value back to the caller and end the function's execution.
Functions without a return statement implicitly return `None`.

```python
def area_rectangle(width, height):
    return width * height

result = area_rectangle(5, 3)
print(result)  # 15
```

## Lambda Functions
Lambda functions are small anonymous functions defined with the `lambda` keyword.
They are useful for short, single-expression callbacks.

```python
square = lambda x: x * x
print(square(6))  # 36
```

## `map` Function
`map` applies a function to every item of an iterable, returning a lazy iterator of results.
Combine `map` with `list()` to materialize the transformed values.

```python
numbers = [1, 2, 3, 4]
doubles = list(map(lambda n: n * 2, numbers))
print(doubles)  # [2, 4, 6, 8]
```

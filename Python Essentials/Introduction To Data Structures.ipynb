```python
# Sets: unordered collections of unique items
tags_post1 = {"python", "coding", "tutorial"}
tags_post2 = {"python", "data", "tutorial"}

common_tags = tags_post1 & tags_post2
all_tags = tags_post1 | tags_post2

print("Post 1 tags:", tags_post1)
print("Common tags:", common_tags)
print("All unique tags:", all_tags)
```

```python
# Nested data structures: lists within dictionaries
course = {
    "title": "Python Programming",
    "students": ["Emma", "Liam", "Olivia"],
    "grades": [88, 92, 85]
}

print("Course:", course["title"])
print("First student:", course["students"][0])
print("Average grade:", sum(course["grades"]) / len(course["grades"]))
```

```python
# List comprehensions: creating lists with concise syntax
numbers = [1, 2, 3, 4, 5, 6]

squares = [n ** 2 for n in numbers]
evens = [n for n in numbers if n % 2 == 0]

print("Original:", numbers)
print("Squares:", squares)
print("Even numbers:", evens)
```

```python
# Task: Create a list of numbers from 1 to 10. Use list comprehension to create a new list containing only odd numbers, and another list with each number multiplied by 3.
```

```python
# Dictionary comprehensions: creating dictionaries efficiently
students = ["Alice", "Bob", "Charlie"]
scores = [95, 87, 92]

grade_book = {student: score for student, score in zip(students, scores)}
high_achievers = {k: v for k, v in grade_book.items() if v >= 90}

print("Grade book:", grade_book)
print("High achievers:", high_achievers)
```

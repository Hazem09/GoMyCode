## Exercise 6: Basic Statistics
Create an array `np.array([12, 15, 20, 22, 18, 17])` and calculate its mean, standard deviation, minimum, maximum, and sum. Print each value with a descriptive label.

Example structure:
```python
scores = np.array([12, 15, 20, 22, 18, 17])
mean_value = scores.mean()
std_value = scores.std()
min_value = scores.min()
max_value = scores.max()
sum_value = scores.sum()
print("Mean:", mean_value)
print("Standard deviation:", std_value)
print("Minimum:", min_value)
print("Maximum:", max_value)
print("Sum:", sum_value)
```

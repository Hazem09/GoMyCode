## Lists
Lists are ordered, mutable collections that can store items of different data types.
They support indexing, slicing, and a rich set of methods for adding, removing, and reorganizing elements.

```python
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print(fruits[1])  # banana
```

## Tuples
Tuples are ordered, immutable sequences, making them ideal for fixed collections of items.
They use parentheses and can be unpacked easily into separate variables.

```python
coordinates = (12.5, 9.8)
x, y = coordinates
print(f"x: {x}, y: {y}")
```

## Sets
Sets are unordered collections of unique elements, useful for membership tests and removing duplicates.
They support mathematical operations like union, intersection, and difference.

```python
student_ids = {101, 102, 103}
student_ids.add(104)
print(102 in student_ids)  # True
```

## Dictionaries
Dictionaries store key-value pairs, allowing fast lookups by key instead of by position.
Keys must be immutable, and methods like `.get()` help provide default values safely.

```python
student = {"name": "Amina", "age": 20, "major": "Biology"}
student["gpa"] = 3.6
print(student.get("age", "Unknown"))
```

# Handling Errors with try, except, and finally

## `try` and `except`

`try` lets you test code that might cause an error. `except` catches the error so your program does not crash.

```python
# A working example of try and except

number_text = "hello"

try:
    number = int(number_text)
    print("The number is:", number)
except ValueError:
    print("That was not a valid number.")
```

## `try`, `except`, and `finally`

`finally` runs no matter what happens. It is useful for cleanup, like closing a file or printing a message at the end.

```python
# A working example of try, except, and finally

text = "5"

try:
    number = int(text)
    print("Number:", number)
except ValueError:
    print("That was not a valid number.")
finally:
    print("This message runs no matter what.")
```
